```python
def load_by_expId_envId(load_dir, expId, envId):
    file_path = load_dir + "/" + str(expId) +"_SngleModel_env_+" +str(envId)+"_reward_logs.npy"
    print("file path = " + file_path)
    reward_log = np.load(file_path)
    return reward_log
```

```python
def load_reward_by_envId(load_dir, exp_num, envId):
    res = []
    for expId in range(exp_num):
#         print(expId)
        cur_load = load_by_expId_envId(load_dir , expId, envId)
        res.append(cur_load)
    return res
```

```python
env0_log = load_reward_by_envId(load_dir, exp_num, envId=0)
env1_log = load_reward_by_envId(load_dir, exp_num, envId=1)
```

```python
def load_all(load_dir, exp_num, env_num):
    all_reward = []
    for envId in range(env_num):
        env_log = load_reward_by_envId(load_dir, exp_num, envId)
        all_reward.append(env_log)
        
    return all_reward
```

```python
all_reward_log = load_all(load_dir, exp_num, env_num)
```

```python

# def plot_single_env_reward(rewards):
#     window_size = 3  
  
#     # 使用NumPy的convolve函数进行移动平均，注意'valid'模式会返回比原数组小的结果  
#     # 为了保持数组大小一致，我们可以使用'full'模式并手动裁剪结果，或者使用'same'模式（但'same'模式在边界处会进行填充）  
#     # 这里我们使用'same'模式，并除以窗口大小来得到平均值  
#     #     smoothed_rewards = np.convolve(rewards, np.ones(window_size)/window_size, mode='same')  
#     # 定义指数移动平均的平滑系数（alpha），接近1时平滑效果较弱，接近0时平滑效果较强  
#     alpha = 0.5
  
#     # 初始化EMA数组，与原始数据相同长度，但第一个值通常设为原始数据的第一个值  
#     smoothed_rewards = np.zeros_like(rewards)  
#     smoothed_rewards[0] = rewards[0]  
  
#     # 计算EMA  
#     for t in range(1, len(rewards)):  
#         smoothed_rewards[t] = alpha * rewards[t] + (1 - alpha) * smoothed_rewards[t-1] 
      
#     # 绘制原始数据和平滑后的数据  
#     plt.figure(figsize=(10, 6))  
#     plt.plot(rewards, label='Original Rewards')  
#     plt.plot(smoothed_rewards, label='Smoothed Rewards', linestyle='--')  
#     plt.legend()  
#     #     plt.title('Smoothing Rewards with Moving Average')  
#     plt.xlabel('Index')  
#     plt.ylabel('Reward Value')  
#     plt.grid(True)  
#     plt.show()

# def plot_all(all_rewards):
#     for i in range(len(all_reward_log)):
#         env_log = np.array(all_reward_log[i])
#         rewards = np.mean(np.mean(env_log, axis=2),axis=0)
#         plot_single_env_reward(rewards)
        
import numpy as np  
import matplotlib.pyplot as plt  
  
def plot_single_env_reward(rewards, title=None):  
    alpha = 0.5  
    smoothed_rewards = np.zeros_like(rewards)  
    smoothed_rewards[0] = rewards[0]  
    for t in range(1, len(rewards)):  
        smoothed_rewards[t] = alpha * rewards[t] + (1 - alpha) * smoothed_rewards[t-1]  
  
    # 找到最高点和最后收敛的值  
    max_orig_reward = np.max(rewards)  
    max_smoothed_reward = np.max(smoothed_rewards)  
    last_orig_reward = rewards[-1]  
    last_smoothed_reward = smoothed_rewards[-1]  
  
    print(f"Highest Original Reward: {max_orig_reward}")  
    print(f"Highest Smoothed Reward: {max_smoothed_reward}")  
    print(f"Last Original Reward: {last_orig_reward}")  
    print(f"Last Smoothed Reward: {last_smoothed_reward}")  
  
    plt.figure(figsize=(10, 6), dpi=300)  # 设置更高的DPI（但这主要用于保存图像）  
    plt.plot(rewards, label='Original Rewards')  
    plt.plot(smoothed_rewards, label='EMA Smoothed Rewards', linestyle='--')  
    plt.legend()  
    plt.xlabel('Index')  
    plt.ylabel('Reward Value')  
    plt.grid(True)  
    if title:  
        plt.title(title)  
    plt.show()  

def plot_all(all_rewards):  
    num_envs = len(all_rewards)  
    fig, axs = plt.subplots(nrows=num_envs, figsize=(10, 6 * num_envs), sharex=True, dpi=300)  # DPI设置对显示影响不大，但可用于保存  
  
    for i, env_log in enumerate(all_rewards):  
        env_log = np.array(env_log)  
        rewards = np.mean(np.mean(env_log, axis=2), axis=0)  
  
        # 绘制和打印每个环境的奖励  
        axs[i].plot(rewards, label='Original Rewards')  
        smoothed_rewards = np.zeros_like(rewards)  
        smoothed_rewards[0] = rewards[0]  
        for t in range(1, len(rewards)):  
            smoothed_rewards[t] = 0.5 * rewards[t] + 0.5 * smoothed_rewards[t-1]  
        axs[i].plot(smoothed_rewards, label='EMA Smoothed Rewards', linestyle='--')  
  
        # 打印信息  
        max_orig_reward = np.max(rewards)  
        max_smoothed_reward = np.max(smoothed_rewards)  
        last_orig_reward = rewards[-1]  
        last_smoothed_reward = smoothed_rewards[-1]  
  
        print(f"Environment {i+1}:")  
        print(f"  Highest Original Reward: {max_orig_reward}")  
        print(f"  Highest Smoothed Reward: {max_smoothed_reward}")  
        print(f"  Last Original Reward: {last_orig_reward}")  
        print(f"  Last Smoothed Reward: {last_smoothed_reward}")  
  
        axs[i].set_title(f'Environment {i+1} Rewards')  
        axs[i].set_xlabel('Index')  
        axs[i].set_ylabel('Reward Value')  
        axs[i].legend()  
        axs[i].grid(True)  
  
    plt.tight_layout()  
    plt.show()
```

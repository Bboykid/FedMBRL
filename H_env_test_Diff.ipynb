```python
from sb3_contrib import TRPO
from Client_diff import FRLClient
from Agent import SB3Agent
import copy
from stable_baselines3.common.evaluation import evaluate_policy
import torch
from H_Envs.pendulum import PendulumEnv
import gymnasium as gym
from gymnasium.wrappers import TimeLimit
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
def evaluate_clients_models(clients, device='cuda' if torch.cuda.is_available() else 'cpu'):
    evaluation_results = {i: [] for i in range(len(clients))}

    for i, client in enumerate(clients):
        client.model.to(device)  # 将模型移动到指定设备
        for j, other_client in enumerate(clients):
            inputs = other_client.test_X.astype(np.float32).values
            labels = other_client.test_y.astype(np.float32).values
            
            inputs = torch.tensor(inputs).to(device)
            labels = torch.tensor(labels).to(device)

            with torch.no_grad():
                outputs = client.model(inputs)
                loss = mean_squared_error(outputs.cpu().numpy(), labels.cpu().numpy())  # 将张量移回 CPU 进行计算

            evaluation_results[i].append(loss)
            print(f"Client {i} evaluated on Client {j}'s data - MSE Loss: {loss:.4f}")

    return evaluation_results
```

```python
# 初始化
TRPO_model = TRPO.load("models/Pendulmn_TRPO_Policy")

timesteps_real_per_round = 500
timesteps_fc_per_round = timesteps_real_per_round * 30
epoch_per_round = 100
CLIENTS_NUM = 3
rounds_num = 10
batch_size_env_model = 128
Global_RL = TRPO_model

gravitys = [8.0, 10.0, 12.0]
real_envs = []

# 初始化真实环境
for gravity in gravitys:
    env = TimeLimit(PendulumEnv(g=gravity), max_episode_steps=200)
    real_envs.append(env)

# 记录每个 round 的数据
round_train_loss = []
round_test_avg = []
round_evaluation_results = []

for round_idx in range(rounds_num):
    env_models = []
    Clients = []
    
    for i in range(CLIENTS_NUM):
        policy_net = Global_RL
        agent = SB3Agent(policy_net)
        client = FRLClient(real_envs[i], agent, lr=3e-4, hidden_size=256, device=device)
        Clients.append(client)
        env_model = copy.deepcopy(client.model)
        env_models.append(env_model)
    
    train_loss_list = []
    test_avg_list = []

    for client_idx in range(len(Clients)):
        print('------------------------------')
        print("client: " + str(client_idx))
        # update policy
        # train prediction models
        Clients[client_idx].learn(timesteps_real_per_round, epoch_per_round, batch_size_env_model)
        train_loss_list.append(Clients[client_idx].train_loss_list)
        test_avg_list.append(Clients[client_idx].test_avg_list)
    
    evaluation_results = evaluate_clients_models(Clients)

    round_train_loss.append(train_loss_list)
    round_test_avg.append(test_avg_list)
    round_evaluation_results.append(evaluation_results)
```

```python
import pickle

# 创建一个字典保存所有数据
data_to_save = {
    'round_train_loss': round_train_loss,
    'round_test_avg': round_test_avg,
    'round_evaluation_results': round_evaluation_results
}

# 将数据保存到文件中
with open('Diff_Env_training_data_test2.pkl', 'wb') as f:
    pickle.dump(data_to_save, f)

print("Data saved successfully.")
```

```python
def calculate_mean_and_variance(data, client_idx):
    # Initialize lists to store all losses for each dataset
    losses = []
    num_datasets = len(data[0][client_idx])

    # Gather losses for each dataset across all rounds
    for i in range(num_datasets):
        losses.append([])

    for round_data in data:
        for i in range(num_datasets):
            losses[i].append(round_data[client_idx][i])

    # Calculate mean and variance for each dataset
    means = [np.mean(loss) for loss in losses]
    variances = [np.var(loss) for loss in losses]
#     print(losses)

    return means, variances
```

```python
def plot_mean_and_variance(round_evaluation_results):
    # Number of clients and datasets
    num_clients = len(round_evaluation_results[0])
    num_datasets = len(round_evaluation_results[0][0])

    # Initialize storage for all means and variances
    all_means = []
    all_variances = []

    # Calculate means and variances for each client
    for client_idx in range(num_clients):
        means, variances = calculate_mean_and_variance(round_evaluation_results, client_idx)
        all_means.append(means)
        all_variances.append(variances)
        # Print means and variances
        print(f"Client {client_idx}:")
        for i in range(num_datasets):
            print(f"  Dataset {i}: Mean = {means[i]:.6f}, Variance = {variances[i]:.6f}")

    # Plot the results
    x = np.arange(num_datasets)
    plt.figure(figsize=(12, 8))

    for client_idx in range(num_clients):
        means = all_means[client_idx]
        variances = all_variances[client_idx]
        std_devs = np.sqrt(variances)
        plt.errorbar(x, means, yerr=std_devs, label=f'Client {client_idx}', 
                     capsize=5, marker='o', linestyle='-', linewidth=2)

    plt.title('Mean and Variance of Losses for Each Client')
    plt.xlabel('Dataset Index')
    plt.ylabel('Loss')
    plt.xticks(x)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_loss_curves_and_statistics(loss_list):
    """
    绘制loss曲线和统计信息图像。

    参数:
    loss_list (list of list of float): 每个元素是一个模型在每个epoch的loss列表。
    """
    # 绘制每个模型的loss曲线
    plt.figure(figsize=(10, 6))
    for idx, loss in enumerate(loss_list):
        plt.plot(loss, label=f'Model {idx+1}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Loss Curves')
    plt.legend()
    plt.grid(True)
    plt.show()

    # 计算每个模型最后一个epoch的loss值
    last_epoch_losses = [loss[-1] for loss in loss_list]

    # 计算最后一个epoch的平均值和方差
    mean_last_epoch_loss = np.mean(last_epoch_losses)
    std_last_epoch_loss = np.std(last_epoch_losses)

    # 绘制最后一个epoch的loss值，平均值和方差
    plt.figure(figsize=(10, 6))
    plt.plot(last_epoch_losses, 'o-', label='Last Epoch Loss')
    plt.axhline(mean_last_epoch_loss, color='r', linestyle='--', label=f'Mean: {mean_last_epoch_loss:.2f}')
    plt.fill_between(range(len(last_epoch_losses)), mean_last_epoch_loss - std_last_epoch_loss, mean_last_epoch_loss + std_last_epoch_loss, color='r', alpha=0.2, label=f'Std Dev: {std_last_epoch_loss:.2f}')
    plt.xlabel('Model Index')
    plt.ylabel('Loss')
    plt.title('Last Epoch Losses with Mean and Std Dev')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_log = np.load("0_test_loss_logs.npy")
```

```python
# 创建一个图形窗口
plt.figure(figsize=(10, 6))

# 绘制每个模型在所有数据集上的loss
for model_idx in range(test_log.shape[1]):  # 遍历每个模型
    # 对每个模型，在所有数据集上进行平均
#     avg_loss = np.mean(test_log[:, model_idx, :], axis=1)
    avg_loss = test_log[:, model_idx, 0]
    plt.plot(avg_loss, label=f'Model {model_idx}')

# 设置图形的标签和标题
plt.xlabel('Round')
plt.ylabel('Average Loss')
plt.title('Model Loss Across Rounds')
plt.legend()

# 显示图形
plt.show()
```

```python
exp_id = 0
test_log = np.load(str(exp_id) +"_test_loss_logs.npy")
reward_log =np.load(str(exp_id) +"_reward_logs.npy")



# 创建一个图形窗口
plt.figure(figsize=(10, 6))

# 绘制每个模型在所有数据集上的loss
for model_idx in range(test_log.shape[1]):  # 遍历每个模型
    # 对每个模型，在所有数据集上进行平均
    avg_loss = np.mean(test_log[:, model_idx, :], axis=1)
    plt.plot(avg_loss, label=f'Model {model_idx}')

# 设置图形的标签和标题
plt.xlabel('Round')
plt.ylabel('Average Loss')
plt.title('Model Loss Across Rounds')
plt.legend()

# 显示图形
plt.show()

# 创建一个图形窗口
plt.figure(figsize=(10, 6))

# 绘制每个模型在所有数据集上的loss
for model_idx in range(test_log.shape[1]):  # 遍历每个模型
    # 对每个模型，在所有数据集上进行平均
    avg_loss = np.mean(reward_log, axis=1)
    plt.plot(avg_loss, label=f'Model {model_idx}')

# 设置图形的标签和标题
plt.xlabel('Round')
plt.ylabel('Reward')
plt.title('Rewards Rounds')
plt.legend()

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_log = np.load("0_test_loss_logs.npy")
```

```python
# 
plt.figure(figsize=(10, 6))

# loss
for model_idx in range(test_log.shape[1]):  # 
    # 
    avg_loss = np.mean(test_log[:, model_idx, :], axis=1)
    plt.plot(avg_loss, label=f'Model {model_idx}')

# 
plt.xlabel('Round')
plt.ylabel('Average Loss')
plt.title('Model Loss Across Rounds')
plt.legend()

# 
plt.show()
```

```python
# 
plt.figure(figsize=(10, 6))

model_idx=0

# loss
for client_idx in range(test_log.shape[1]):  # 
    # 
#     avg_loss = np.mean(test_log[:, model_idx, :], axis=1)
    avg_loss = test_log[:, model_idx,client_idx]
    plt.plot(avg_loss, label=f'dataset {client_idx}')

# 
plt.xlabel('Round')
plt.ylabel('Average Loss')
plt.title(f'Model Loss Across Rounds model {model_idx}')
plt.legend()

# 
plt.show()
```
